## 1. 데이터 준비
* 토큰화가 잘 되어 있는 filtered 데이터 사용

```python
import pandas as pd
data_filename = './data/Korean_movie_reviews_2016_filtered.csv'
data_df = pd.read_csv(data_filename)
print(data_df.head())
```

```python
data_df.info()
```

```python
data_df.dropna(inplace=True)
data_df.info()
```

```python
# review만 모아서 review별 토큰 리스트로 변환 : review가 Object 타입이므로 str로 변환 후 split
review_list =[review.split() for review in list(data_df.review)]
print(review_list[:2])

# review_list = []
# for review in list(data_df.review):
#     review_list.append(review.split())
```

## 1. Word2Vec 활용 영화 리뷰 워드 임베딩
* https://radimrehurek.com/gensim/models/word2vec.html

### Skipgram, negative=10 인 경우

```python
# Word2Vec 모델 생성 및 학습 : window=3, min_count=3
from gensim.models import Word2Vec

model = Word2Vec(review_list, vector_size=100, sg=1, negative=10, window=3, min_count=5)
print(model)
```

```python
# 단어의 임베딩 벡터 확인
model.wv['이정재']
```

```python
# 단어의 임베딩 벡터 차원 확인
len(model.wv['이정재'])
```

```python
# 두 단어 간 유사도 확인
model.wv.similarity('이정재', '정우성')
```

```python
# 특정 단어와 유사한 단어 추출
model.wv.most_similar('이정재', topn=20)
```

```python
model.wv.most_similar('재밌', topn=50)
```

```python
model_sg_n10 = model
```

## 2. FastText 활용 영화 리뷰 워드 임베딩
* https://radimrehurek.com/gensim/models/fasttext.html

```python
# FastText 모델 생성 및 학습
# window=3, min_count=3, min_n=2, max_n=2
from gensim.models import FastText
model_ft = FastText(review_list, vector_size=100, sg=1, negative=10, window=3, min_count=3, min_n=2, max_n=2)
```

```python
# 특정 단어와 유사한 단어 추출 : 이정재
model_ft.wv.most_similar('이정재', topn=20)
```

```python
# corpus에 없는 단어 확인 : 우주평화 
'우주평화' in model_ft.wv.key_to_index
```

```python
# corpus에 없는 단어의 임베딩 벡터 확인 
model_ft.wv['우주평화']
```

```python
# corpus에 없는 단어와 유사한 단어추출 
model_ft.wv.most_similar('우주평화', topn=20)
```
