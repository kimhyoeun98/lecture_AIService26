```python
# 데이터 준비 (로딩)
data_filename = './data/daum_movie_review.csv'
input_filename = 'daum_movie_review.csv'
import pandas as pd
movie_df = pd.read_csv(data_filename)
#movie_df.head()
review_list = list(movie_df.review)
print(type(review_list))
print(review_list[:10])
# 리뷰별로 넣고 pos한 이후에 합치기. pos_list += 새로운 리뷰 이렇게 쓰던가 pos_list.extend(새로운 리뷰) 
```

```python
from konlpy.tag import Okt
tagger = Okt()

noun_list = []

my_stopwords = ['돈', '!', '?', '헐', '사람', '진짜', '보고', '마지막', '시간', '그냥', '정도','영화','생각','마동석','윤계상','내용','연기','감동','정말','광주','마블','감독']

noun_list = [noun for review in review_list for noun in tagger.nouns(review) if noun not in my_stopwords and len(noun) >1]

print(len(noun_list))

print(noun_list[:])
```

```python
from collections import Counter
const_counter = Counter(noun_list)

len(const_counter)

most_freq_list = const_counter.most_common(20)
```

```python
x_list = [ noun for noun, count in most_freq_list ]
y_list = [ count for noun, count in most_freq_list ]
print('x 데이터 : ', x_list)
print('y 데이터 : ', y_list)

from matplotlib import font_manager, rc
font_path = "c:/Windows/Fonts/malgun.ttf"
font_name = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font_name)

import matplotlib.pyplot as plt

plt.barh(x_list[::-1], y_list[::-1])

# 그래프 정보 추가 (제목, 레이블)
plt.title('다음 영화 리뷰 분석')
plt.xlabel('빈도수')
plt.ylabel('키워드')

result_filename = './result/' + input_filename.replace('.csv', '_graph_김효은.jpg')
plt.savefig(result_filename)
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

font_path = "c:/Windows/fonts/malgun.ttf"

const_wc = WordCloud(
    font_path = font_path,
    width = 800,
    height = 600,
    max_words = 50,
    background_color = 'white'
                    )

const_wc = const_wc.generate_from_frequencies(const_counter)

plt.imshow(const_wc)
plt.axis('off')

result_filename = './result/' + input_filename.replace('.csv', '_wordcloud_김효은.jpg')
plt.savefig(result_filename)
```
